## Daily Tesla Stock data from the last two years

### Munging for stock volume by month

```python
# Importing the data
# Tesla stock daily closing stats
TS_2Years = pd.read_csv('../Data/HistoricalQuotes _2_years.csv')
```

```python
# Looking at data
print(len(TS_2Years))
print(TS_2Years.ftypes)
```

```python
# Looking at data
TS_2Years.head()
```

```python
# Removing first row data as it's not in the right format.
TS_2Years.drop(TS_2Years.index[:1], inplace = True)
```

```python
# Checking it worked
TS_2Years.head()
```

```python
# Converting to datetime, pd.to_datetime
TS_2Years.index =  pd.to_datetime(TS_2Years['date'], format='%Y/%m/%d')
```

```python
# looking at data
TS_2Years.head()
```

```python
# Looking at data
TS_2Years.ftypes
```

```python
# Looking at data
TS_2Years.columns
```

```python
# Changing 'volume' from a object to float
TS_2Years['volume'] = TS_2Years['volume'].astype(float)
```

```python
# Sorting and adding you stock volume by year and month
df2 = TS_2Years.groupby([list(TS_2Years.index.year),list(TS_2Years.index.month)])['volume'].sum()
```

```python
# Volume of shares of Tesla traded permonth, for the two years I have data for.
df2
```

```python
# Another way of doing it
# Volume of shares of Tesla traded permonth, for the two years I have data for.
TS_2Years['volume'].sum()
TS_2Years.resample('M')['volume'].sum()
```

### Munging for number of days the stock was traded a month, that I have data for

```python
# Counts the number of days the stock was traded a month
Times_a_month = TS_2Years.groupby([list(TS_2Years.index.year),list(TS_2Years.index.month)]).count()
```
