# Web Scrapping
Using lession 5.07 as reference

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Scrapping for the first 40 data points

```python
# going by 40 results a page because I could not get more than 41 rows for some reason, the site can return up to a 100
url = "https://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&CIK=0001318605&type=&dateb=&owner=exclude&count=40.html"
```

```python
response = requests.get(url)
```

```python
print('Status Code: ',response.status_code)
# Ex., 200 = OK, 400 = Bad Request, 403 = Forbidden, 404 = Not Found.
```

```python
html = response.text
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
soup
```

```python
table = soup.find("table", {"class":"tableFile2"})
```

```python
tr = table.find_all("tr")
df1 = []
for i in range(1, (len(tr))):
    result = {}
    td = tr[i].find_all('td')
    
    if td:
        result['Filings'] = td[0].text
        result['Date'] = td[3].text
    if len(result) == 2:
        df1.append(result)
```

## Coverting the dictionary to a pandas dataframe

```python
df1 = pd.DataFrame(df1, columns=(['Filings', 'Date']))
```

```python
df1
```
